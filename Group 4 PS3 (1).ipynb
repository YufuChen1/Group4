```python
"Question 1"
"a)"
# import all the libraries needed
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt

# set N to be 1000 or 10000 and the step to be 100
N= 10000
step = 100

# numberIndex is to control the iteration from 0 to N
numberIndex = 0

# the list const[] to restore beta0, x1[] to restore beta1
const = []
x1 = []

# while loop for iteration
while numberIndex < N:
    sigma_e = 5

# generate a list of random numbers which obey the normal distribution
    x2 = np.random.normal(0,1,step)
    e = np.random.normal(0,sigma_e,step)

    y1 = 2 + 0.5*x2 + e

# do the ols and store the beta0 and beta1 in const[] and x1[] separatedly
    X = sm.add_constant(x2)
    ols1 = sm.OLS(y1,X).fit()
    const.append(ols1.params[0])
    x1.append(ols1.params[1])
    numberIndex += step

# use the kdeplot of seaborn to draw density figure
sns.kdeplot(const)
sns.kdeplot(x1)

# use plt.show() to show the figure
plt.show()
print(ols1.summary())
```

```python
"Question 4"

"a)"
def gini(x):
  total = 0
  for i, xi in enumerate(x[:-1], 1):
    total += np.sum(np.abs(xi - x[i:]))
  return total / (len(x) ** 2 * np.mean(x))

def y_dist():
  sigma = 0.8
  mu = 7.5
  log_y = np.random.normal() * sigma + mu # reparameterise
  y = 2 ** log_y
  return y

y_sim = []
for _ in range(100000):
  y_sim.append(y_dist())

y_sim = np.array(y_sim)
print('mean', y_sim.mean())
print('var', y_sim.var())
print('gini', gini(y_sim) )
```

```python
"b)"
def c_dist():
  y = y_dist()
  c = y ** 0.8 + 0.5 * y + 500
  return c
c_sim = []
for _ in range(100000):
 c_sim.append(c_dist())
c_sim = np.array(c_sim)
print('gini', gini(c_sim))
print('mean', c_sim.mean())
print('var', c_sim.var())
```
