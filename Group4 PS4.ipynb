```python
"Question 3"
"a)"
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from scipy import optimize
"Define utility function"
def u(c, theta=1.5):
    return (c**(1-theta))/(1-theta)
beta = 0.96
"define obejective function"
def objective(x, y0=10, y1=5, a0=0, beta=0.96, r=(1/beta) - 1, theta=1.5):
    c0, c1 = x
    a1 = (y0 + a0 - c0)
    return -u(c0, theta) - beta*u(c1, theta)
"define budget constraints"
def budget(x, y0=10, y1=5, a0=0, beta=0.96, r=(1/beta) - 1):
    c0, c1 = x
    a1 = (y0 + a0 - c0)
    return [(1+r)*a1 + y1 - c1]
"initial guess for c0 and c1"
x0 = [5, 5]
"solution"
sol = optimize.minimize(objective, x0, args=(10, 5, 0, 0.96, (1/beta) - 1, 1.5), method='SLSQP', constraints={'type': 'eq', 'fun': budget})
c0_opt, c1_opt = sol.x
print("optimal c0 = ",c0_opt)
print("optimal c1 = ",c1_opt)
```

```python
"b)"
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Set parameters
theta = 1.5
beta = 0.96
r = 1/beta - 1
y = np.array([5, 10, 15, 0])
a0 = 0

# Define utility function
def u(c):
    return (c**(1-theta))/(1-theta)

# Define objective function
def objective(x):
    c0, c1, c2, c3, a1, a2, a3 = x
    return -u(c0) - beta*u(c1) - beta**2*u(c3) - beta**3*u(c2)


# Define budget constraint functions
def budget0(x):
    c0, c1, c2, c3, a1, a2, a3 = x
    return c0 + a1 - a0 - y[0]

def budget1(x):
    c0, c1, c2, c3, a1, a2, a3 = x
    return c1 + a2 - (1+r)*a1 - y[1]

def budget2(x):
    c0, c1, c2, c3, a1, a2, a3 = x
    return c2 + a3 - (1+r)*a2 - y[2]

def budget3(x):
    c0, c1, c2, c3, a1, a2, a3 = x
    return c3 - (1+r)*a3 - y[3]

# Set initial guess for variables
x0 = [1, 1, 1, 1, 1, 1, 1]

# Set bounds for variables
bounds = [(0, None), (0, None), (0, None), (0, None), (None, None), (None, None), (None, None)]

# Set constraints
cons = [{'type': 'eq', 'fun': budget0},
        {'type': 'eq', 'fun': budget1},
        {'type': 'eq', 'fun': budget2},
        {'type': 'eq', 'fun': budget3}]

# Minimize the negative of the objective function
result = minimize(objective, x0, bounds=bounds, constraints=cons)

# Extract optimal values
c0, c1, c2, c3, a1, a2, a3 = result.x

# Print optimal values
print("Optimal consumption levels:")
print(c0)
print(c1)
print(c2)
print(c3)
```

```python
"d)"
import numpy as np
from scipy.optimize import minimize_scalar

# Define the utility function and parameters
theta = 1.5
beta = 0.96
r = (1 / beta) - 1

def u(c):
    return (c**(1-theta)) / (1 - theta)

# Define the income realizations and probabilities
y1l = 2.5
y1h = 7.5
pl = 0.5
ph = 0.5

# Define the objective function to be minimized
def obj_func(c0):
    a1 = (c0 + 10) / (1 + r)
    E_u = pl*u((1+r)*a1+y1l) + ph*u((1+r)*a1+y1h)
    return -(u(c0) + beta*E_u)

# Find the optimal c0 using scalar optimization
res = minimize_scalar(obj_func, method='bounded', bounds=(0, 10))

# Extract the optimal consumption and saving decisions
c0 = res.x
a1 = (c0 + 10) / (1 + r)
c1l = (1 + r)*a1 + y1l - a1
c1h = (1 + r)*a1 + y1h - a1

# Print the results
print("Optimal consumption in period 0:", c0)
print("Optimal saving in period 1:", a1)
print("Optimal consumption in period 1 (low income):", c1l)
print("Optimal consumption in period 1 (high income):", c1h)

# Plot the income and consumption paths for both income realizations
import matplotlib.pyplot as plt

y1_vals = [y1l, y1h]
c1_vals = [c1l, c1h]

fig, ax = plt.subplots()
ax.plot([0, 1], [10, y1_vals[0]], label='Low income')
ax.plot([0, 1], [10, y1_vals[1]], label='High income')
ax.plot([0, 1], [c0, c1_vals[0]], label='Consumption (low income)')
ax.plot([0, 1], [c0, c1_vals[1]], label='Consumption (high income)')
ax.set_xlabel('Period')
ax.set_ylabel('Income / Consumption')
ax.legend()
plt.show()
```
